```python
import torch
import matplotlib.pyplot as plt

x = torch.arange(0,10000,1)
# 周期项
y1 = torch.sin(torch.pi*x/15)
# 趋势项
y2 = 0.003*x+0.3
# 扰动项
y3 = torch.normal(0,0.1,y1.shape)


plt.plot(x[:100],(y1+y2+y3)[:100])
# plt.plot(x,y1+y2+y3)
plt.show()
```

```python
import numpy as np

np.savetxt('mydata.csv', np.column_stack((x,y1,y2,y3)), fmt='%.3f', delimiter=",",header='oder_data',comment='')

# data = np.loadtxt('test.csv')
```
