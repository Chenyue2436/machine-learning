```python
!wget -O simhei.ttf "https://www.wfonts.com/download/data/2014/06/01/simhei/chinese.simhei.ttf"
import matplotlib.pyplot as plt
import matplotlib as mpl
zhfont = mpl.font_manager.FontProperties(fname='simhei.ttf')
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
#打印结果
import numpy as np

setting = "SCINet_ETTh1_ftS_sl48_ll48_pl24_lr0.009_bs16_hid4.0_s1_l3_dp0.25_invFalse_itr0"
np.set_printoptions(precision=3) #设置小数位置为3位

input = np.load('./exp/ett_results/'+setting+'/input.npy')
print("输入的数据为：{}".format(input[0,:,0]))

pred = np.load('./exp/ett_results/'+setting+'/pred_scales.npy')
print("预测的结果为：{}".format(pred[0,:,0]))

true = np.load('./exp/ett_results/'+setting+'/true_scales.npy')
print("真实的结果为：{}".format(true[0,:,0]))

metrics = np.load('./exp/ett_results/'+setting+'/metrics_scales.npy')
print("mae, mse, rmse, mape, mspe分别为：{}".format(metrics[:-1]))
```

```python
#画图
x1 = np.arange(72)
x2 = np.arange(47,72)
y_t = np.concatenate((input[0,:,0],true[0,:,0]))
y_p = np.concatenate(([input[0,-1,0]],pred[0,:,0]))

plt.plot(x1,y_t,color='b')
plt.plot(x2,y_p,color='r')
plt.xlabel(u'时间',fontproperties=zhfont)
plt.ylabel(u'油温',fontproperties=zhfont)
plt.legend(('GroundTruth', 'Prediction'), loc='upper right')
plt.title(u'SCINet 2017/10/24的预测结果',fontproperties=zhfont)

plt.savefig('./SCINet_ETTh1_S_48_48_24.png')
plt.show()
```
