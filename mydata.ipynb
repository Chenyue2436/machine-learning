```python
import torch
import matplotlib.pyplot as plt

x = torch.arange(0,10000,1)
# 周期项
y1 = torch.sin(torch.pi*x/15)
# 趋势项
y2 = 0.003*x+0.3
# 扰动项
y3 = torch.normal(0,0.1,y1.shape)


plt.plot(x[:100],(y1+y2+y3)[:100])
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt

x = torch.arange(0,10000,1)
# 周期项
y1 = torch.sin(2*torch.pi*x/7) + 2*torch.cos(2*torch.pi*x/30) + torch.exp(3*torch.sin(2*torch.pi*x/365))

# 趋势项
y2 = torch.zeros(x.shape)

# 扰动项
# y3 = torch.normal(0,0.1,y1.shape)
y3 = torch.zeros(x.shape)


plt.plot(x[:1000],(y1+y2+y3)[:1000])
plt.show()
```

```python
import numpy as np

y = y1+y2+y3
np.savetxt('mydata_t.csv', np.column_stack((x,y1,y2,y3,y)), fmt='%.3f', delimiter=",",header='date,y1,y2,y3,y',comments='')

# data = np.loadtxt('test.csv')
```
